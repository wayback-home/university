```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
result = []

for page in range(1, 39):
    stock_url = "https://finance.naver.com/sise/sise_market_sum.naver?&page=%d" % page


    html = urllib.request.urlopen(stock_url)
    soupStock = BeautifulSoup(html, "html.parser",from_encoding="euc-kr")
    tag_tbody = soupStock.find("tbody")
    for stock in tag_tbody.find_all("tr",onmouseover="mouseOver(this)"):
        # if len(stock) <= 11:
        #     break
        stock_td = stock.find_all("td")
        stock_title = stock_td[1].string
        stock_price = stock_td[2].string
        stock_yesterday = stock_td[3].find("span").string.strip("\n""\t")
        stock_updown = stock_td[4].find("span").string.strip("\n""\t")
        stock_perPrice = stock_td[5].string
        stock_total = stock_td[6].string
        stock_value = stock_td[7].string
        stock_foreign = stock_td[8].string
        stock_trades = stock_td[9].string
        stock_PER = stock_td[10].string
        stock_ROE = stock_td[11].string
        
        result.append(
            [stock_title]
            + [stock_price]
            + [stock_yesterday]
            + [stock_updown]
            + [stock_perPrice]
            + [stock_total]
            + [stock_value]
            + [stock_foreign]
            + [stock_trades]
            + [stock_PER]
            + [stock_ROE]
        )

```

```python
stock_td = pd.DataFrame(result, columns=("종목명","현재가","전일비","등략률","액면가","시가총액","상장주식수","외국인비율","거래량","PER","ROE"))
```
